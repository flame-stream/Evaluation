# Rates

```python
%matplotlib inline
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
plt.rcParams["figure.figsize"] = [16, 14] 
```

```python
items = pd.DataFrame(columns=["workers", "rate", "good", "all"])

for workers in os.listdir(os.fsencode("rates")):
    w = os.fsdecode(workers)
    for rate in os.listdir(os.fsencode("rates/" + w)):
        r = os.fsdecode(rate)
        all = np.int32(0)
        good = np.int32(0)
        for node in os.listdir(os.fsencode("rates/" + w + "/" + r)):
            n = os.fsdecode(node)
            with open(os.fsencode("rates/" + w + "/" + r + "/" + n)) as f:
                data = f.read().strip()
                all += np.int32(data.split("/")[0])
                good += np.int32(data.split("/")[1])
        df =  pd.DataFrame(data = {'workers' : [np.int32(w)], 'rate' : [np.int32(r)], 'good' : [good], 'all' : [all]})       
        items = pd.concat([items, df])
```

```python
items['coef'] = items['all'] / items['good']
```

```python
workers = np.array([1, 2, 4, 5, 7, 8, 10])
rates = np.array([70, 80, 90, 100, 125, 150, 200, 250, 300])
```

```python
X, Y = np.meshgrid(rates, workers)
Z = np.zeros((len(workers), len(rates)))
for i, w in enumerate(workers):
    for j, r in enumerate(rates):
        if (w == 1):
            Z[i][j] = 1
        else:
            Z[i][j] = items[(items['rate'] == r) & (items['workers'] == w)]['coef'][0]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('interarrival (ms)')
ax.set_ylabel('workers')
ax.set_zlabel('ratio')
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False)
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.savefig("fig.pdf", format='pdf')
```
