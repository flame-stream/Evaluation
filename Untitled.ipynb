# FlameStream single worker, remote bench

```r
require(data.table)
```

```r
fs.1w <- tail(scan("fs-1w.data", sep=","), -200)
fs.2w <- tail(scan("fs-2w.data", sep=","), -200)
fs.5w <- tail(scan("fs-5w.data", sep=","), -200)
fs.7w <- tail(scan("fs-7w.data", sep=","), -200)
fs.10w <- tail(scan("fs-10w.data", sep=","), -200)
```

```r
fl.1w <- tail(scan("fl-1w.data", sep=","), -200)
fl.2w <- tail(scan("fl-2w.data", sep=","), -200)
fl.5w <- tail(scan("fl-5w.data", sep=","), -200)
```

```r
summary(fs.1w)
summary(fs.2w)
summary(fs.5w)
summary(fs.7w)
summary(fs.10w)
```

```r
summary(fl.1w)
summary(fl.2w)
summary(fl.5w)
```

```r
quntiles <- seq(1, length(fs.1w)) / length(fs.1w)
```

```r
quantile(fs.1w, p=0.95)
quantile(fs.2w, p=0.95)
quantile(fs.5w, p=0.95)
quantile(fs.7w, p=0.95)
quantile(fs.10w, p=0.95)
```

```r
quantile(fl.1w, p=0.95)
quantile(fl.2w, p=0.95)
quantile(fl.5w, p=0.95)
```

```r
plot(quntiles, sort(fs.1w), type="l", col="purple", xlim = c(0.4, 0.8), ylim=c(1e7, 1.5e8), xlab="quantile", ylab="latency")
lines(quntiles, sort(fs.2w), col="blue")
lines(quntiles, sort(fs.5w), col="green")
lines(quntiles, sort(fs.7w), col="orange")
lines(quntiles, sort(fs.10w), col="red")
legend(0.6, 8e7, title = "FlameStream", legend=c("1 worker", "2 workers", "5 workers", "7 workers", "10 workers"), col=c("purple", "blue", "green", "orange", "red"), lty=1, cex=0.8)

lines(quntiles, sort(fl.1w), lty="dotted", col="purple")
lines(quntiles, sort(fl.2w), lty="dotted", col="blue")
lines(quntiles, sort(fl.5w), lty="dotted", col="green")
legend(0.4, 1.5e8, title = "Flink", legend=c("1 worker", "2 workers", "5 workers", "7 workers", "10 workers"), col=c("purple", "blue", "green", "orange", "red"), lty="dotted", cex=0.8)

```

```r
plot(quntiles, sort(fs.1w), type="l", col="purple", xlim = c(0.95, 1), xlab="quantile", ylab="latency")
lines(quntiles, sort(fs.2w), col="blue")
lines(quntiles, sort(fs.5w), col="green")
lines(quntiles, sort(fs.7w), col="orange")
lines(quntiles, sort(fs.10w), col="red")
legend(0.95, 9e8, title = "FlameStream", legend=c("1 worker", "2 workers", "5 workers", "7 workers", "10 workers"), col=c("purple", "blue", "green", "orange", "red"), lty=1, cex=0.8)

lines(quntiles, sort(fl.1w), lty="dotted", col="purple")
lines(quntiles, sort(fl.2w), lty="dotted", col="blue")
lines(quntiles, sort(fl.5w), lty="dotted", col="green")
legend(0.964, 8e8, title = "Flink", legend=c("1 worker", "2 workers", "5 workers", "7 workers", "10 workers"), col=c("purple", "blue", "green", "orange", "red"), lty="dotted", cex=0.8)
```
